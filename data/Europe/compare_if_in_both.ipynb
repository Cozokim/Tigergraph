```python
import pandas as pd
csv1 = r"C:\Users\JulienRigot\OneDrive - LIS Data Solutions\Escritorio\code_GORDIAS\base de datos graph\Tigergraph\data\Europe\biggest_europe_cities_without_CODE.csv"
csv2= r"C:\Users\JulienRigot\OneDrive - LIS Data Solutions\Escritorio\code_GORDIAS\base de datos graph\Tigergraph\data\Europe\conexiones_biggest_europe_cities_without_code_dailymovement80%.csv"
# Read the datasets from CSV files with a specified encoding
try:
    df1 = pd.read_csv(csv1, encoding='utf-8')  # Try UTF-8 encoding
except UnicodeDecodeError:
    df1 = pd.read_csv(csv1, encoding='ISO-8859-1')  # Fallback to ISO-8859-1

try:
    df2 = pd.read_csv(csv2, encoding='utf-8')  # Try UTF-8 encoding
except UnicodeDecodeError:
    df2 = pd.read_csv(csv2, encoding='ISO-8859-1')  # Fallback to ISO-8859-1
```

```python

# Extract the unique cities from the first dataset
cities_dataset1 = set(df1['City'])

# Extract unique cities from the second dataset (both origin and destination)
cities_dataset2 = set(df2['CODE_Origin']).union(set(df2['CODE_Destination']))

# Find cities that are not in both datasets
not_in_dataset1 = cities_dataset2 - cities_dataset1
not_in_dataset2 = cities_dataset1 - cities_dataset2

# Display the results
print("Cities in dataset 2 but not in dataset 1:")
print(not_in_dataset1)

print("\nCities in dataset 1 but not in dataset 2:")
print(not_in_dataset2)
```

```python
# Remove rows containing cities not in the other dataset
df1_filtered = df1[df1['City'].isin(cities_dataset2)]
df2_filtered = df2[df2['CODE_Origin'].isin(cities_dataset1) & df2['CODE_Destination'].isin(cities_dataset1)]

# Save the filtered DataFrames to new CSV files (optional)
df1_filtered.to_csv('filtered_dataset1.csv', index=False)
df2_filtered.to_csv('filtered_dataset2.csv', index=False)

# Display filtered DataFrames
print("\nFiltered dataset 1:")
print(df1_filtered)

print("\nFiltered dataset 2:")
print(df2_filtered)
```
