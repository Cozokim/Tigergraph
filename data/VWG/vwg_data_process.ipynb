```python
df_nodes = pd.read_csv(csv_nodes)
df_conexions = pd.read_csv(csv_aristas)
```

```python
df_conexions
```

```python
df_conexions.columns = df_conexions.iloc[1]  # La ligne 2 devient les noms des colonnes
df_conexions = df_conexions[2:]  # Conserve les lignes à partir de l'index 2

# Réinitialise l'index
df_conexions = df_conexions.reset_index(drop=True)
```

```python
df_conexions.columns = df_conexions.iloc[0]  # La première ligne devient les noms des colonnes
df_conexions = df_conexions[1:]  # Conserve les lignes à partir de l'index 1

# Réinitialise l'index
df_conexions = df_conexions.reset_index(drop=True)
```

```python
df_conexions
```

```python
df_conexions = df_conexions.drop(df_conexions.columns[0], axis=1)


```

```python
df_conexions
```

```python
# Fusionner df_conexions avec df_nodes pour obtenir la colonne 'City_Origin'
df_conexions = df_conexions.merge(df_nodes[['MasterCode', 'City']], left_on='MasterCode_Origin', right_on='MasterCode', how='left')
df_conexions = df_conexions.rename(columns={'City': 'City_Origin'}).drop(columns='MasterCode')

# Fusionner df_conexions avec df_nodes pour obtenir la colonne 'City_Destination'
df_conexions = df_conexions.merge(df_nodes[['MasterCode', 'City']], left_on='MasterCode_Destination', right_on='MasterCode', how='left')
df_conexions = df_conexions.rename(columns={'City': 'City_Destination'}).drop(columns='MasterCode')
```

```python
df_conexions_filtered = df_conexions[['City_Origin', 'City_Destination']]
```
